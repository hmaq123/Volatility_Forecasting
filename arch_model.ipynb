```python
import numpy as np
import pandas as pd
import yfinance as yf
from arch import arch_model
```

```python
df = yf.download('^GSPC', start = '2022-01-01', end = '2025-09-20')

df['returns'] = df['Close'].pct_change()*100 #daily % returns
returns = df['returns'].dropna()

model = arch_model(returns, vol = 'ARCH', p = 1) # Call the model
results = model.fit(disp = 'off') # Train the model

results.summary()
```

```python
forecast = results.forecast(horizon = 5) #Forecast 5 days ahead
predicted_variance = forecast.variance

predicted_volatility = np.sqrt(predicted_variance) #vol= sqrt of variance
predicted_volatility
```

```python
predicted_volatility = [0.984373,1.117502,1.159814,1.174032,1.178888]
predicted_avg_vol = sum(predicted_volatility)/len(predicted_volatility)
predicted_avg_vol
```

```python
real_df = yf.download('^GSPC', start= '2025-09-19', end = '2025-09-27')
real_df['returns'] = real_df['Close'].pct_change()*100
real_df = real_df.dropna()
realized_vol = real_df['returns'].std()*np.sqrt(5) # Predicting the vol 
realized_vol
```

```python
print("ARCH Model Predicted Volatility:", predicted_avg_vol)
print("ARCH Model Actual Volatility:", realized_vol)
```
